```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
import random
```

```python
df = pd.DataFrame(np.around(np.random.uniform(low=0, high=10, size=(300,4)), decimals=1), columns=list('ABCD'))
df['B'] = df['B']*100
df['C'] = df['C']*1000
```

```python
the_list = ['x', 'y', 'z'] * 100
```

```python
df['x_axis'] = [random.choice(x) for x in the_list]
```

```python
df
```

```python
sns.set_style(style='whitegrid')
sns.set_context(context='poster')

sns.boxplot(x='x_axis', y='A', data=df)
sns.despine()
plt.show()

sns.boxplot(x='x_axis', y='B', data=df)
sns.despine()
plt.show()

sns.boxplot(x='x_axis', y='C', data=df)
sns.despine()
plt.show()
```

# Oh no! All different scales and not faceted! 
Solution: melt data, use sns.factorplot()

```python
df_melt = pd.melt(df, id_vars='x_axis')
df_melt
```

```python
sns.factorplot(x='x_axis', y='value', data=df_melt, kind='box', 
               col='variable', col_wrap=2)
plt.show()
```
